```python
# Gather Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression

# Random state
randomState = 10
```

```python
# Gather Data
data = pd.read_csv('student_scores.csv')

X = data['Hours'].values
y = data['Scores'].values
```

```python
# Initial Parameters
weight = 0.0
bias = 0.0
lr = .001
epochs = 100
n = len(X)
mse = []
```

```python
# Training loop
for epoch in range(epochs):
  y_pred = [weight * x + bias for x in X]
  error = y_pred - y
  mse.append((error**2).mean())

  dw = (2/n) * (X.T @ error)
  db = (2/n) * error.sum()

  weight -= lr * dw
  bias -= lr* db

plt.figure(figsize = (10, 5))

plt.subplot(1, 2, 1)
plt.scatter(X, y, color="blue", label = "Data")
plt.plot(X, [weight * x + bias for x in X], color = "red", label = "Regression line")
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title("Linear Regression with Gradient Descent")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot([e for e in range(epochs)], mse)
plt.xlabel('epoch')
plt.ylabel('MSE')
plt.title("Gradient Descent Diverging")

plt.tight_layout()
plt.show()
```
