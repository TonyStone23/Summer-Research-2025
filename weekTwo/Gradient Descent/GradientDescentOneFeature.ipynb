```python

# Gather Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression as mr

# Random state
randomState = 10
```

```python
X, y = mr(50, 1, noise = 15, random_state = randomState)
plt.figure(figsize = (4, 4))
plt.scatter(X, y)
plt.show()
```

```python
X = pd.DataFrame(X).values
y = pd.DataFrame(y).values
print(X.shape)
print(y.shape)
```

```python
# Initialize
weight = 0.0
bias = 0.0
lr = .01
epochs = 500
n = len(X)
mse = []
```

```python
# Training loop
for epoch in range(epochs):
  y_pred = np.array([weight * x + bias for x in X]).reshape(-1, 1)
  error = y_pred - y
  mse.append((error**2).mean())

  dw = (2/n) * (X.T @ error)
  db = (2/n) * error.sum()

  weight -= lr * dw
  bias -= lr * db
```

```python
plt.subplot(1, 2, 1)
plt.scatter(X, y, color="blue", label = "Data")
plt.plot(X, np.array([weight * x + bias for x in X]).reshape(-1, 1), color = "red", label = "Regression line")
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Linear Regression with Gradient Descent")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot([e for e in range(epochs)], mse)
plt.xlabel('epoch')
plt.ylabel('MSE')
plt.title("Gradient Descent Diverging")

plt.tight_layout()
plt.show()
```
